```python
#counterfactual prompts
#Sentiment signals generation using DeepSeek-V3 for counterfactual prompting approach
#to generate the output file 'sentiment_counterfactual.csv'

def get_sentiment_Counterfactual(symbol, *texts):
    texts = [text for text in texts if text != 0]
    num_text = len(texts)
    text_content = " ".join([f"### News to Stock Symbol -- {symbol}: {text}" for text in texts])

    conversation = [
    {
        "role": "system",
        "content": (
            "You are a financial analyst. For each headline, consider what the opposite situation would be (a counterfactual), "
            "and then reason whether the actual news is positive, negative, or neutral. "
            "Give a sentiment score from 1 to 5 (1 = very negative, 5 = very positive). "
            "Respond only with numeric scores."
        )
    },
    {
        "role": "user",
        "content": "News to Stock Symbol -- AAPL: Apple launches new iPhone model"
    },
    {
        "role": "assistant",
        "content": "If Apple did not launch a new model, the market might think innovation is slowing. So the actual news is somewhat positive. Score: 4"
    },
    {
        "role": "user",
        "content": "News to Stock Symbol -- AAPL: Apple delays software update"
    },
    {
        "role": "assistant",
        "content": "If there were no delay, it would indicate smooth operations. The delay implies execution issues. Score: 2"
    },
    {
        "role": "user",
        "content": text_content
    }
    ]

    return conversation

```

```python
def build_counterfactual_prompt(symbol, texts):
    """
    Builds a counterfactual reasoning prompt for a list of financial headlines.
    """
    prompt = (
        "You are a financial analyst. For each headline, consider what the opposite situation would be "
        "(a counterfactual), and then reason whether the actual news is positive, negative, or neutral.\n"
        "Assign a sentiment score from 1 to 5 (1 = very negative, 5 = very positive).\n"
        "Respond only with: Score: X for each headline.\n\n"
    )

    for i, text in enumerate(texts, 1):
        prompt += f"{i}. News to Stock Symbol -- {symbol}: {text.strip()}\n"

    prompt += "\nReturn sentiment scores only as:\nScore: 3\nScore: 2\nScore: 5 ..."
    return prompt
```

```python
def process_csv_Counterfactual(input_path, output_path, batch_size=5, chunk_size=1000):
    import re
    reader = pd.read_csv(input_path, chunksize=chunk_size)
    all_results = []

    for chunk in reader:
        chunk = chunk.dropna(subset=['Lsa_summary'])
        chunk['sentiment'] = np.nan

        for i in range(0, len(chunk), batch_size):
            batch = chunk.iloc[i:i + batch_size]
            texts = batch['Lsa_summary'].tolist()
            symbol = batch['symbol'].iloc[0] if 'symbol' in batch.columns else "AAPL"

            try:
                prompt = build_counterfactual_prompt(symbol, texts)
                print(f"\nðŸ§  Calling LLM for {len(texts)} headlines...")

                chat_completion = openai.chat.completions.create(
                    model='deepseek-ai/DeepSeek-V3',
                    messages=[{"role": "user", "content": prompt}],
                    temperature=0,
                    max_tokens=500,
                    stream=False,
                )

                # Validate response
                if not chat_completion or not chat_completion.choices:
                    raise ValueError("âŒ No response or empty choices from LLM")

                content = chat_completion.choices[0].message.content

                if not content:
                    raise ValueError("âŒ Response content is empty")

                print("ðŸ” LLM Output:\n", content)

                # Flexible sentiment score extraction
                scores = re.findall(r"(?:Score|Sentiment|Rating)[:=]?\s*([1-5])", content, re.IGNORECASE)
                print("ðŸŽ¯ Extracted Scores:", scores)

                sentiments = [int(score) for score in scores[:len(texts)]]

                if len(sentiments) < len(texts):
                    print(f"âš ï¸ Warning: Only extracted {len(sentiments)} scores for {len(texts)} texts.")

            except Exception as e:
                print(f"âŒ Error: {e}")
                sentiments = [np.nan] * len(texts)

            chunk.loc[batch.index, 'sentiment'] = sentiments

        all_results.append(chunk)

    final_df = pd.concat(all_results)
    final_df.to_csv(output_path, index=False)
    print(f"\nâœ… Saved results to {output_path}")
```

```python
if __name__ == "__main__":
    input_file = '/content/aapl_filtered.csv'  # Replace with your input file path
    output_file = '/content/sentiment_counterfactual.csv'
    process_csv_Counterfactual(input_file, output_file, batch_size=5, chunk_size=1000)
```
