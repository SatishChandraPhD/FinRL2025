#Merging data
Key Considerations
Primary Key:
Since the analysis is date-based, the primary key should be the date.

If both files contain a stock symbol column (e.g., "tic"), use the composite key ["date", "tic"].

In this case, since we are only dealing with AAPL, the trading data uses only trading days while the sentiment data includes all days (trading plus weekends/holidays).

Merging Strategy:
Use a left join with the trading data as the base. This way, the final dataset will contain only trading days. For dates without a corresponding sentiment value, you will have NaN, the data has been handled by forward filling  

```python
import pandas as pd

# Load datasets
trading_df = pd.read_csv("/content/aapl_trading_data_4Merging.csv")
sentiment_df = pd.read_csv("/content/aapl_sentiment_data_4merging.csv")

# Convert date columns to datetime and remove timezone info
trading_df['date'] = pd.to_datetime(trading_df['date']).dt.tz_localize(None)
sentiment_df['date'] = pd.to_datetime(sentiment_df['date']).dt.tz_localize(None)

# Ensure tic column exists for merging
if 'tic' not in trading_df.columns:
    trading_df['tic'] = 'AAPL'
if 'tic' not in sentiment_df.columns:
    sentiment_df['tic'] = 'AAPL'

# Merge the two dataframes using a left join on ['date', 'tic']
merged_df = pd.merge(trading_df, sentiment_df, on=['date', 'tic'], how='left')

# Check the merged DataFrame
print("Merged DataFrame Shape:", merged_df.shape)
print(merged_df.head())

# Forward-fill missing sentiment data (if any)
merged_df['sentiment'] = merged_df['sentiment'].ffill()
```

```python
merged_df.to_csv('aapl_trading_sentiment_data.csv', index=False)
```

```python
import pandas as pd

# Load the two CSV files
trading_df = pd.read_csv("/content/aapl_trading_data_4Merging.csv")
sentiment_df = pd.read_csv("/content/aapl_sentiment_data_4merging.csv")

# Convert the date columns to datetime and remove timezone info if needed
trading_df['date'] = pd.to_datetime(trading_df['date']).dt.tz_localize(None)
sentiment_df['date'] = pd.to_datetime(sentiment_df['date']).dt.tz_localize(None)

# If the symbol column isn't present, add it (since you are dealing only with AAPL)
if 'tic' not in trading_df.columns:
    trading_df['tic'] = 'AAPL'
if 'tic' not in sentiment_df.columns:
    sentiment_df['tic'] = 'AAPL'

# Perform an outer join on the composite key ['date', 'tic']
merged_df_all = pd.merge(trading_df, sentiment_df, on=['date', 'tic'], how='outer')

# Sort by date so we have a proper time series order
merged_df_all = merged_df_all.sort_values(by='date').reset_index(drop=True)

# Now you have rows for both trading days and the extra dates from the sentiment data

# Option: If the sentiment information is in a column (e.g., 'sentiment_deepseek'),
# you may want to club consecutive dates by aggregating the sentiment over a window.
# Define the window size (n days)
n = 3  # for example, a 3-day window; you can adjust this value later

# To perform a rolling aggregation, set the date as index (ensure index is sorted).
merged_df_all.set_index('date', inplace=True)

# Compute a rolling average for the sentiment column.
# Here, we assume that missing sentiment values (NaN) remain NaN — if desired, you can fill them first.
merged_df_all['sentiment_rolling_avg'] = merged_df_all['sentiment'].rolling(window=n, min_periods=1).mean()

# Reset the index to bring the 'date' back as a column.
merged_df_all.reset_index(inplace=True)

# Inspect the merged DataFrame
print("Merged DataFrame Shape:", merged_df_all.shape)
print(merged_df_all.head(10))
```

```python
merged_df_all.to_csv('chainofthought_aapl_trading_sentiment_data_all_days.csv', index=False)
```
